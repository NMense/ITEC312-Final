```python
import sqlite3

class Team:
    def __init__(self, team_name, player_count, school):
        self.team_name = team_name
        self.player_count = player_count
        self.school = school

class Sport:
    def __init__(self, sport_name):
        self.sport_name = sport_name

class Match:
    def __init__(self, sport, team1, team2, score, location):
        self.sport = sport
        self.team1 = team1
        self.team2 = team2
        self.score = score
        self.location = location

def save_match_to_db(match):
    connection = sqlite3.connect('matches.db')
    cursor = connection.cursor()

    cursor.execute('''CREATE TABLE IF NOT EXISTS matches 
                      (sport_name TEXT, team1_name TEXT, team2_name TEXT, score TEXT, location TEXT)''')

    cursor.execute('''INSERT INTO matches VALUES (?, ?, ?, ?, ?)''', 
                   (match.sport.sport_name, match.team1.team_name, match.team2.team_name, match.score, match.location))

    connection.commit()
    connection.close()

def search_matches_by_team(team_name):
    connection = sqlite3.connect('matches.db')
    cursor = connection.cursor()

    cursor.execute('''SELECT * FROM matches WHERE team1_name = ? OR team2_name = ?''', (team_name, team_name))
    matches = cursor.fetchall()

    connection.close()
    return matches

def menu():
    while True:
        print("\nMenu:")
        print("1. Add a Team")
        print("2. Add a Sport")
        print("3. Add a Match")
        print("4. Search Matches by Team")
        print("5. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            team_name = input("Enter team name: ")
            player_count = input("Enter player count: ")
            school = input("Enter school: ")
            team = Team(team_name, player_count, school)
            # Save team to database or perform other actions if needed

        elif choice == '2':
            sport_name = input("Enter sport name: ")
            sport = Sport(sport_name)
            # Save sport to database or perform other actions if needed

        elif choice == '3':
            # Get sport, team1, team2, score, location inputs and create a Match object
            sport_name = input("Enter sport name: ")
            team1_name = input("Enter team 1 name: ")
            team2_name = input("Enter team 2 name: ")
            score = input("Enter score: ")
            location = input("Enter location: ")

            sport = Sport(sport_name)
            team1 = Team(team1_name, 0, "")
            team2 = Team(team2_name, 0, "")
            match = Match(sport, team1, team2, score, location)
            save_match_to_db(match)

        elif choice == '4':
            team_to_search = input("Enter team name to search: ")
            matches = search_matches_by_team(team_to_search)
            if matches:
                print("Matches found:")
                for match in matches:
                    print(match)
            else:
                print("No matches found for the given team.")

        elif choice == '5':
            break

        else:
            print("Invalid choice. Please enter a valid option.")

if __name__ == "__main__":
    menu()
```
